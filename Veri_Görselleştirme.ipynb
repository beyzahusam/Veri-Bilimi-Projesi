50 farklı startup şirketinin çeşitli harcamalarını ve kârlılıklarını içeren veri setinin analizi yapılacaktır.
Öncelikle gerekli kütüphaneler eklenmeli ve veri seti okunmalıdır.

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\beyza\\OneDrive\\Masaüstü\\VeriBilimi\\50_Startups.csv")
```

R&D harcaması ile kâr arasındaki ilişkiyi scatter plot ile gösterme:

```python
plt.figure(figsize=(6,4))
plt.scatter(df["R&D Spend"], df["Profit"], color="blue", alpha=0.6)
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D Harcaması vs Kâr")
plt.show()
```

Grafikten görülebileceği gibi R&D harcaması arttıkça şirketin kârı da artmaktadır. Burada güçlü bir korelasyon görülmektedir. Şirketin başta düşük harcamalarla yüksek kâr elde ettiği, sonradan ise kârın lineer arttığı çıkarılabilir.

Yönetim harcaması ile kâr arasındaki ilişkiyi scatter plot ile gösterme:


```python
plt.figure(figsize=(6,4))
plt.scatter(df["Administration"], df["Profit"], color="green", alpha=0.6)
plt.xlabel("Administration Spend")
plt.ylabel("Profit")
plt.title("Yönetim Harcaması vs Kâr")
plt.show()
```

Grafiğe göre yönetim harcaması ve kâr arasında güçlü bir korelasyon bulunmamaktadır. Aralarında doğrudan bir ilişki yoktur ve harcamaya göre kârın durumu hakkında belirli bir tahmin yapmak zordur. Yönetim harcamasının kâr üzerinde tek başına belirleyici bir faktör olmadığı görülebilir.

Eyaletlere göre ortalama kârları bar chart ile gösterme:


```python
state_profit = df.groupby("State")["Profit"].mean()
plt.figure(figsize=(6,4))
state_profit.plot(kind="bar", color="orange", edgecolor="black")
plt.ylabel("Ortalama Kâr")
plt.title("Eyaletlere Göre Ortalama Kâr")
plt.show()
```

Grafikte verilerde yer alan 3 eyaletin ortalama kârı gösterilmektedir. En çok kâr Florida, en az ise California'dadır.


R&D, Yönetim ve Pazarlama harcamalarının dağılımını boxplot ile gösterme:


```python
plt.figure(figsize=(8,5))
df[["R&D Spend", "Administration", "Marketing Spend"]].boxplot()
plt.title("Harcamaların Dağılımı (Boxplot)")
plt.ylabel("Harcama Miktarı")
plt.show()
```

Grafikte belirli türdeki harcamaların ortalaması, 1. ve 3. çeyrekleri,maksimum ve minimum değerleri görülebilir. Grafikten çıkarılabileceği üzere ortalama en az harcama miktarına sahip olan tür R&D, en çok ortalama harcama miktarına sahip ise pazarlamadır. Harcamalarda R&D'nin daha kontrollü ve sınırlı, yönetim harcamalarının orta düzeyde ve dar bir aralıkta değer aldığı, pazarlama harcamalarının ise çok değişken olduğu görülmektedir.
